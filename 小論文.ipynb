```python
import pandas as pd
chart = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRux-hrYQTYaqNm_FuW6vn6UXeFm3jNOI49y6TvwwjnJhih4_7uU2sB_ntFTBJO36ZR2Fa-LhtGeNzh/pub?output=csv")
chart
```

```python
year = ['1959前','1960-1979','1980-1989','1990-1999','2000-2009','2010後']
bornlist = list(chart["出生年"])
count = {}
for i in year:
    count.update({i : bornlist.count(i)})
borncount = {x:y for x,y in count.items() if y!=0}
borncount
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(10,10))
plt.pie(borncount.values(), labels = borncount.keys(), autopct='%1.1f%%', startangle=140, textprops={'fontsize':25, 'color':'w'})
plt.axis("equal")
plt.title("出生年圓餅統計圖",fontsize=25)
plt.xticks(fontsize=20)
plt.legend(fontsize=20,loc=7)
plt.show()
```

```python
import matplotlib.patches as mpatches
colors=["red", "orange", "blue","orange" ,"orange","blue","blue","orange","red","orange" ]
datanum = len(chart)
frqtdict = {}
def frqt(chartname):
    wordname = chartname.strip("下列用詞﹐你曾否聽過?").replace("[", " ").replace("]", " ").strip()
    word = list(chart[chartname])
    wordfrqt = (word.count("聽過1~2次") * 1+ word.count("經常聽到") * 2 + word.count("從未聽過") * 0 ) / datanum
    frqtdict.update({wordname : wordfrqt})
frqt("下列用詞﹐你曾否聽過? [郭]")
frqt("下列用詞﹐你曾否聽過? [這我一定吉]")

frqt("下列用詞﹐你曾否聽過? [確認過眼神]")
frqt("下列用詞﹐你曾否聽過? [可撥]")
frqt("下列用詞﹐你曾否聽過? [不要再打了]")
frqt("下列用詞﹐你曾否聽過? [skr]")
frqt("下列用詞﹐你曾否聽過? [母湯]")
frqt("下列用詞﹐你曾否聽過? [8+9]")
frqt("下列用詞﹐你曾否聽過? [想不到吧]")
frqt_pd = pd.Series(frqtdict).to_frame() 
plt.figure(figsize=(10,8))
plt.bar(range(len(frqtdict)), list(frqtdict.values()), color=colors)
plt.xticks(range(len(frqtdict)), list(frqtdict.keys()), ha='right', rotation=45,fontsize=13)
plt.yticks(fontsize=20)
red_patch = mpatches.Patch(color='red', label='口耳相傳')
blue_patch = mpatches.Patch(color='blue', label='傳統媒體')
orange_patch = mpatches.Patch(color='orange', label="社群網路")
plt.legend(handles=[red_patch, blue_patch, orange_patch], fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel("分數(0~2)",fontsize=17)
plt.title("流行語傳唱程度長條圖", fontsize=20)
plt.savefig("C:\\Users\\ck402\\OneDrive\\小論文\\流行語傳唱程度長條圖", bbox_inches = "tight")
plt.show()
```

```python
comefrom = ['傳統媒體(ej: 新聞、報紙、 雜誌)', '網路社群', '同儕、 朋友', '家人', '從未聽過']

orgdict= {}
def org(chartname):
    wordlist = []
    wordname = chartname.strip("下列用詞﹐你是從哪裡聽到的?").replace("[", " ").replace("]", " ").strip()
    for i in comefrom:
        wordlist.append(list(chart[chartname]).count(i))
    orgdict.update({wordname : wordlist})
org("下列用詞﹐你是從哪裡聽到的? [郭]")
org("下列用詞﹐你是從哪裡聽到的? [這我一定吉]")

org("下列用詞﹐你是從哪裡聽到的? [確認過眼神]")
org("下列用詞﹐你是從哪裡聽到的? [可撥]")
org("下列用詞﹐你是從哪裡聽到的? [不要再打了]")
org("下列用詞﹐你是從哪裡聽到的? [skr]")
org("下列用詞﹐你是從哪裡聽到的? [母湯]")
org("下列用詞﹐你是從哪裡聽到的? [8+9]")
org("下列用詞﹐你是從哪裡聽到的? [想不到吧]")
orgdict_pd = pd.DataFrame(orgdict, index=comefrom)
orgdict_pd_T = orgdict_pd.T
orgtalk_df = orgdict_pd_T["同儕、 朋友"].add(orgdict_pd_T["家人"], fill_value=0).to_frame()
orgdict_pd_T["口耳相傳"] = orgtalk_df
del orgdict_pd_T["同儕、 朋友"]
del orgdict_pd_T["家人"]
orgdict_pd_T.plot(kind="bar", figsize=(13,6), width=0.8)
plt.xticks(fontsize=15,rotation = 45)
plt.yticks(fontsize=15)
plt.legend(fontsize=11, loc=9)
plt.title("流行語來源分析長條圖",fontsize=20)
plt.ylabel("選填人數",fontsize=15)
plt.savefig("C:\\Users\\ck402\\OneDrive\\小論文\\流行語來源分析長條圖", bbox_inches = "tight")
plt.show()
```

```python
datanum = len(chart)
feldict = {}
def fel(chartname):
    wordname = chartname.strip("當他人使用這些流行語時，你的感覺是什麼?").replace("[", " ").replace("]", " ").strip()
    word = list(chart[chartname])
    wordfel = (word.count("非常良好") * 4 + word.count("良好") * 3 + word.count("厭惡") * 2 + word.count("非常厭惡") * 1) / datanum
    feldict.update({wordname : wordfel})
fel("當他人使用這些流行語時，你的感覺是什麼? [郭]")
fel("當他人使用這些流行語時，你的感覺是什麼? [這我一定吉]")

fel("當他人使用這些流行語時，你的感覺是什麼? [確認過眼神]")
fel("當他人使用這些流行語時，你的感覺是什麼? [可撥]")
fel("當他人使用這些流行語時，你的感覺是什麼? [不要再打了]")
fel("當他人使用這些流行語時，你的感覺是什麼? [skr]")
fel("當他人使用這些流行語時，你的感覺是什麼? [母湯]")
fel("當他人使用這些流行語時，你的感覺是什麼? [8+9]")
fel("當他人使用這些流行語時，你的感覺是什麼? [想不到吧]")
fel_pd = pd.Series(feldict).to_frame() 
fel_pd.plot(kind="bar", figsize=(8,6))
plt.xticks(range(len(feldict)), list(feldict.keys()), ha='right', rotation=45,fontsize=15)
plt.yticks(range(0,5),fontsize=20)
plt.title("流行語感官程度長條圖",fontsize=20)
plt.ylabel("分數(0~4)",fontsize=20)
plt.savefig("C:\\Users\\ck402\\OneDrive\\小論文\\流行語感官程度長條圖", bbox_inches = "tight")
plt.show()
```
